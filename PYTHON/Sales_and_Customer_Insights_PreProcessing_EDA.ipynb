# STEP 5: Exploratory Data Analysis - EDA

```python
# ------------------------------
#  KPIs (Key Metrics)
# ------------------------------
#print("Total Units Returned:", df['Return'].sum())
# KPI 1: Total Units Returned
units_returned = df.loc[df["Return"].str.lower() == "yes", "Units"].sum()
print("Total Units Returned:", units_returned)
# Sum of Units by Return (Yes/No)
units_by_return = df.groupby("Return")["Units"].sum().reset_index()

print("ðŸ“Š Sum of Units by Return")
print(units_by_return)
print("ðŸ“Š Sum of Revenue by State")
revenue_by_state = df.groupby("State")["Revenue"].sum().reset_index()
print(revenue_by_state)
print("Total Revenue:", df['Revenue'].sum())

print("Total Price:", df['Price'].sum())
print("Total Sales:", df['Revenue'].sum())   # Sales = Revenue here
```

```python
# ------------------------------
#  Revenue by Product
# ------------------------------
plt.figure(figsize=(6,4))
sns.barplot(x="Product", y="Revenue", data=df, estimator=sum)
plt.title("Sum of Revenue by Product")
plt.show()
```

```python
# ------------------------------
#  Revenue by Acquisition Type
# ------------------------------
plt.figure(figsize=(6,6))
df.groupby("Customer Acquisition Type")["Revenue"].sum().plot.pie(
    autopct="%1.1f%%", startangle=90, explode=[0.05]*df["Customer Acquisition Type"].nunique()
)
plt.title("Revenue by Customer Acquisition Type")
plt.ylabel("")
plt.show()
```

```python
# ------------------------------
#  Revenue by State
# ------------------------------
plt.figure(figsize=(8,4))
sns.barplot(x="State", y="Revenue", data=df, estimator=sum)
plt.title("Sum of Revenue by State")
plt.xticks(rotation=45)
plt.show()
```

```python
# ------------------------------
#  Revenue Trend by Year
# ------------------------------
plt.figure(figsize=(6,4))
sns.lineplot(x="Date", y="Revenue", data=df, estimator=sum, marker="o")
plt.title("Revenue by Year")
plt.show()
```

```python
# ------------------------------

#  Return Count by Product
# ------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Product", hue="Return", data=df)
plt.title("Return Count by Product")
plt.show()
```

```python
# ------------------------------
# Customer Satisfaction
# ------------------------------
plt.figure(figsize=(6,4))
sns.countplot(x="Customer Satisfaction", data=df, order=df["Customer Satisfaction"].value_counts().index)
plt.title("Customer Satisfaction Distribution")
plt.show()
```
